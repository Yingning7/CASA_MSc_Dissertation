```python
from pprint import pprint
import sys
sys.path.append('C:/Users/danie/Desktop/yingning')

from data_utils import fetch_selected_dataset, save_reses
from models import tree
```

```python
dataset_configs = [
    (False, 0, False),
    (False, 0, True),
    (True, 0, False),
    (True, 0, True),
    (False, 1, False),
    (False, 1, True),
    (True, 1, False),
    (True, 1, True)
]
model_func = tree
max_depths = [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
folder_name = 'tree_max_depth{max_depth}'
```

```python
for max_depth in max_depths:
    reses = {}
    for dataset_config in dataset_configs:
        res = model_func(fetch_selected_dataset(*dataset_config), max_depth=max_depth)
        reses[str(dataset_config)] = res
        pprint(res['score'])
    save_reses(reses, folder_name.format(max_depth=max_depth))
```
